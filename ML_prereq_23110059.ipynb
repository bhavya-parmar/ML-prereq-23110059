Q9. Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.


```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0.5,100,0.5)

#1
y=x
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = x')
plt.grid(True)
plt.show()


#2
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = x^2')
plt.grid(True)
plt.show()

#3
y=x**3/100
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = x^3/100')
plt.grid(True)
plt.show()

#4
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x)')
plt.grid(True)
plt.show()

#5
y=np.sin(x)/x
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x)/x')
plt.grid(True)
plt.show()

#6
y=np.log(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = log(x)')
plt.grid(True)
plt.show()

#7
y=np.exp(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = e^x')
plt.grid(True)
plt.show()
```

Q10. Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as â€œaâ€, â€œbâ€, â€œcâ€, â€œdâ€, â€œeâ€. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1, 2, size=(20, 5))
df = pd.DataFrame(matrix, columns=['a', 'b', 'c', 'd', 'e'])

print(df)

max_std=-np.inf
max_std_col=''
for i in df.columns:
  if np.std(df[i])>max_std:
    max_std=np.std(df[i])
    max_std_col=i

print("Max STD at column", max_std_col,"=", max_std)

min_mean=min(np.mean(df, axis=1))
min_mean_row=np.argmin(np.mean(df, axis=1))
print(f"Min mean at row {min_mean_row} = {min_mean}")
```

The value in the column â€œgâ€ should be â€œLT8â€ if the value in the column â€œfâ€ is less than 8 and â€œGT8â€ otherwise. Find the standard deviation of the column â€œfâ€ for the rows where the value in the column â€œgâ€ is â€œLT8â€ and â€œGT8â€ respectively.

```python
df['f']=df['a'] + df['b'] + df['c'] + df['d'] + df['e']

df['g']=np.where(df['f']<8, 'LT8', 'GT8')

lt8_count = df[df['g'] == 'LT8'].shape[0]

lt8_std = df[df['g'] == 'LT8']['f'].std()
gt8_std = df[df['g'] == 'GT8']['f'].std()

print(df)
print("No. of rows with LT8 =", lt8_count)
print("STD of rows with LT8 =", lt8_std)
print("SRD of rows with GT8 =", gt8_std)
```

Q12. Write a small piece of code to explain broadcasting in numpy.

```python
''' Numpy Array Broadcasting is the ability of NumPy to treat arrays of different dimensions during arithmetic operations.
In this the array of smaller dimension can be broadcast across the array with larger dimension, ensuring that they have compatible shapes for these operations.'''

#Example 1
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])

c = a + b # Adds 10 to every first column elements, 20 to second column elements and 30 to third column elements

print(c)

#Example 2

a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[10], [20]])

c = a + b # Adds 10 to every first row elements and 20 to second row elements

print(c)
```

Q13. Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
def argmin(arr):
  min_val=arr[0]
  min_idx=0
  for i in range(1, len(arr)):
    if min_val>arr[i]:
      min_val=arr[i]
      min_idx=i
  return min_idx

# Example
arr=np.array([2,3,4,1,5,6,3])
print("Using argmin function = ", argmin(arr))
print("Using numpy argmin = ", np.argmin(arr))
```
